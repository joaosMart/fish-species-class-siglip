```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import glob
import os
import json
import random
from decord import VideoReader, cpu
from collections import defaultdict
import matplotlib.gridspec as gridspec

# First, let's remove any existing hooks
def clear_all_hooks(model):
    """Remove all existing hooks from the model"""
    for module in model.modules():
        module._forward_hooks.clear()
        module._forward_pre_hooks.clear()
        module._backward_hooks.clear()

class TaskLevelOcclusionAnalyzer:
    """
    Task-level occlusion analysis using SVM classification only
    """
    def __init__(self, model, preprocess, device, trained_svm, label_encoder):
        self.model = model  # SigLIP model
        self.preprocess = preprocess
        self.device = device
        self.trained_svm = trained_svm
        self.le = label_encoder

        # Clear any existing hooks
        clear_all_hooks(self.model)

    def extract_siglip_features(self, image):
        """Extract SigLIP features from image"""
        processed_image = self.preprocess(image).unsqueeze(0).to(self.device)
        with torch.no_grad():
            features = self.model.encode_image(processed_image)
            features = F.normalize(features, dim=-1)
        return features.cpu().numpy()

    def tensor_to_pil(self, tensor):
        """Convert tensor back to PIL Image"""
        # Move tensor to CPU first
        tensor = tensor.cpu()

        # Denormalize tensor (assuming ImageNet normalization)
        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
        tensor = tensor * std + mean
        tensor = torch.clamp(tensor, 0, 1)

        # Convert to PIL
        image_np = (tensor.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
        return Image.fromarray(image_np)

    def svm_occlusion_analysis(self, image, window_size=12, stride=4):
        """
        Occlusion analysis using SVM classification confidence
        """
        # Get baseline features and prediction
        baseline_features = self.extract_siglip_features(image)
        baseline_pred = self.trained_svm._predict_proba_lr(baseline_features)[0]
        baseline_confidence = baseline_pred.max()
        baseline_class = self.trained_svm.predict(baseline_features)[0]

        print(f"Baseline prediction: {self.le.classes_[baseline_class]} (confidence: {baseline_confidence:.3f})")

        # Preprocess for occlusion (resize to 224x224)
        processed_image = self.preprocess(image).unsqueeze(0).to(self.device)
        h, w = processed_image.shape[-2:]  # 224, 224

        importance_map_small = np.zeros((h, w))

        print(f"Running task-level occlusion analysis...")

        total_windows = 0
        for y in range(0, h - window_size + 1, stride):
            for x in range(0, w - window_size + 1, stride):
                total_windows += 1

        processed_windows = 0
        for y in range(0, h - window_size + 1, stride):
            for x in range(0, w - window_size + 1, stride):
                # Create occluded image
                occluded_image_tensor = processed_image.clone()
                mean_values = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1).to(self.device)
                occluded_image_tensor[:, :, y:y+window_size, x:x+window_size] = mean_values

                # Convert back to PIL for feature extraction
                occluded_image_pil = self.tensor_to_pil(occluded_image_tensor.squeeze(0))

                # Get features and prediction
                occluded_features = self.extract_siglip_features(occluded_image_pil)
                occluded_pred = self.trained_svm._predict_proba_lr(occluded_features)[0]
                occluded_confidence = occluded_pred.max()

                # Compute confidence drop (higher = more important region)
                confidence_drop = baseline_confidence - occluded_confidence

                # Assign importance to all pixels in this window
                importance_map_small[y:y+window_size, x:x+window_size] += confidence_drop

                processed_windows += 1
                if processed_windows % 50 == 0:
                    print(f"Processed {processed_windows}/{total_windows} windows")

        # Normalize importance map
        importance_map_small = (importance_map_small - importance_map_small.min()) / (importance_map_small.max() - importance_map_small.min() + 1e-8)

        # Resize to match original image dimensions
        orig_h, orig_w = image.size[1], image.size[0]  # PIL size is (width, height)
        importance_map_full = cv2.resize(importance_map_small, (orig_w, orig_h), interpolation=cv2.INTER_LINEAR)

        return importance_map_full, baseline_class, baseline_confidence

    def task_level_occlusion_analysis(self, image, save_path=None, window_size=12, stride=4):
        """
        Run task-level occlusion analysis on the image
        """
        print(f"\n=== Running Task-Level Occlusion Analysis ===")

        # Run SVM analysis
        svm_map, predicted_class, confidence = self.svm_occlusion_analysis(image, window_size=window_size, stride=stride)


        fig, axes = plt.subplots(1, 2, figsize=(12, 6))

        # Original image (left)
        axes[0].imshow(image)
        axes[0].axis('off')

        # SVM importance (right)
        axes[1].imshow(image, alpha=0.7)
        im = axes[1].imshow(svm_map, alpha=0.5, cmap='jet',
                           extent=[0, image.size[0], image.size[1], 0])

        class_name = self.le.classes_[predicted_class]
        axes[1].axis('off')

        # Smaller colorbar
        plt.colorbar(im, ax=axes[1], fraction = 0.025, pad=0.02, label='Confidence Drop')

        plt.tight_layout()

        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Task-level analysis saved to: {save_path}")

        plt.show()

        # Print analysis summary
        print(f"\n=== Analysis Summary ===")
        print(f"SVM importance range: {svm_map.min():.3f} to {svm_map.max():.3f}")

        return {
            'svm_map': svm_map,
            'predicted_class': predicted_class,
            'confidence': confidence,
            'class_name': class_name
        }

def load_data_and_train_svm(data_dir, seed=37):
    """
    Load data and train SVM for occlusion analysis
    """
    print("Loading NPZ files...")

    # Load data
    features_list = []
    labels_list = []

    npz_files = glob.glob(os.path.join(data_dir, "*.npz"))
    print(f"Found {len(npz_files)} NPZ files")

    for npz_file in npz_files:
        try:
            data = np.load(npz_file, allow_pickle=True)
            features = data['averaged_features']
            fish_species = str(data['fish_species'].item())

            if features is not None:
                features_list.append(features)
                labels_list.append(fish_species)
        except Exception as e:
            print(f"Error processing {npz_file}: {e}")
            continue

    # Convert to arrays
    X = np.array(features_list)
    y = np.array(labels_list)

    print(f"Loaded {len(X)} samples")
    print(f"Classes: {np.unique(y)}")

    # Encode labels
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=0.2, random_state=seed, stratify=y_encoded
    )

    print(f"Training set: {len(X_train)} samples")
    print(f"Test set: {len(X_test)} samples")

    # Train SVM with default parameters
    print("Training SVM...")
    svm = LinearSVC(random_state=seed, class_weight='balanced', max_iter=2000)
    svm.fit(X_train, y_train)

    # Evaluate
    train_acc = svm.score(X_train, y_train)
    test_acc = svm.score(X_test, y_test)

    print(f"Training accuracy: {train_acc:.3f}")
    print(f"Test accuracy: {test_acc:.3f}")

    return svm, le, (X_train, X_test, y_train, y_test), X, y

def extract_frame_from_npz(npz_file_path, base_video_dir="/content/drive/MyDrive/Summer 2023 - fish ML"):
    """
    Extract the original frame from video using NPZ file information
    """
    try:
        # Load NPZ file
        npz_data = np.load(npz_file_path, allow_pickle=True)

        # Get frame number
        middle_frame = int(npz_data["middle_frame"].item())

        # Parse filename to get collection_site_id and video_id
        npz_filename = os.path.basename(npz_file_path)
        name_parts = npz_filename.replace('_features.npz', '').split('_')

        collection_site_id = name_parts[0]
        video_id = name_parts[1]

        # Construct video path with correct format
        video_path = f"{base_video_dir}/{collection_site_id}_vid/{video_id}.mp4"

        print(f"Extracting frame {middle_frame} from video: {video_path}")

        # Load video and extract frame
        vr = VideoReader(video_path, ctx=cpu(0))
        frame = vr[middle_frame].asnumpy()

        # Convert to PIL Image
        pil_image = Image.fromarray(frame)

        # Get fish species info
        fish_species = str(npz_data['fish_species'].item())

        print(f"Successfully extracted frame for {fish_species}")
        return pil_image, fish_species, collection_site_id, video_id, middle_frame

    except Exception as e:
        print(f"Error extracting frame from {npz_file_path}: {e}")
        return None, None, None, None, None

def extract_correct_and_incorrect_frames(npz_dir, svm_model, label_encoder, features_array, labels_array,
                                        confidence_threshold=0.6, max_correct=5, max_incorrect=5, seed=42):
    """
    Extract frames from test set with both correct and incorrect predictions
    """
    try:
        # Get test set NPZ files with their indices
        y_encoded = label_encoder.transform(labels_array)
        X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
            features_array, y_encoded, range(len(features_array)),
            test_size=0.2, random_state=seed, stratify=y_encoded
        )

        # Get all NPZ files
        npz_files = [f for f in os.listdir(npz_dir) if f.endswith('.npz')]
        npz_files.sort()  # Ensure consistent ordering

        # Calculate confidence for each test sample
        correct_candidates = []
        incorrect_candidates = []

        print(f"Evaluating {len(X_test)} test samples for correct/incorrect predictions...")

        for i, test_idx in enumerate(idx_test):
            # Get features for this sample
            sample_features = X_test[i].reshape(1, -1)

            # Get prediction probabilities
            pred_proba = svm_model._predict_proba_lr(sample_features)[0]
            confidence = pred_proba.max()
            predicted_class = svm_model.predict(sample_features)[0]
            true_class = y_test[i]

            # Check if confidence meets threshold
            if confidence >= confidence_threshold:
                npz_path = os.path.join(npz_dir, npz_files[test_idx])
                candidate_info = {
                    'npz_path': npz_path,
                    'confidence': confidence,
                    'predicted_class': predicted_class,
                    'true_class': true_class,
                    'predicted_species': label_encoder.classes_[predicted_class],
                    'true_species': label_encoder.classes_[true_class],
                    'test_index': i,
                    'is_correct': predicted_class == true_class
                }

                if predicted_class == true_class:
                    correct_candidates.append(candidate_info)
                else:
                    incorrect_candidates.append(candidate_info)

        print(f"Found {len(correct_candidates)} correct predictions with confidence >= {confidence_threshold}")
        print(f"Found {len(incorrect_candidates)} incorrect predictions with confidence >= {confidence_threshold}")

        # If not enough samples, try lowering threshold
        if len(correct_candidates) < max_correct or len(incorrect_candidates) < max_incorrect:
            print(f"Not enough samples found. Trying lower confidence threshold...")
            if confidence_threshold > 0.5:
                return extract_correct_and_incorrect_frames(
                    npz_dir, svm_model, label_encoder, features_array, labels_array,
                    confidence_threshold=max(0.5, confidence_threshold-0.1),
                    max_correct=max_correct, max_incorrect=max_incorrect, seed=seed
                )

        # Sort by confidence (highest first)
        correct_candidates.sort(key=lambda x: x['confidence'], reverse=True)
        incorrect_candidates.sort(key=lambda x: x['confidence'], reverse=True)

        # Select top samples
        selected_correct = correct_candidates[:max_correct]
        selected_incorrect = incorrect_candidates[:max_incorrect]

        print(f"\nSelected {len(selected_correct)} correct predictions:")
        for i, candidate in enumerate(selected_correct, 1):
            print(f"{i}. {candidate['predicted_species']}: {candidate['confidence']:.3f} (CORRECT)")

        print(f"\nSelected {len(selected_incorrect)} incorrect predictions:")
        for i, candidate in enumerate(selected_incorrect, 1):
            print(f"{i}. {candidate['true_species']} → {candidate['predicted_species']}: {candidate['confidence']:.3f} (INCORRECT)")

        # Extract frames for both correct and incorrect
        all_extracted_frames = []

        print(f"\n=== Extracting Correct Prediction Frames ===")
        for candidate in selected_correct:
            npz_path = candidate['npz_path']
            print(f"Processing {os.path.basename(npz_path)} (confidence: {candidate['confidence']:.3f})")

            image, fish_species, collection_site, video_id, frame_num = extract_frame_from_npz(npz_path)

            if image is not None:
                info = {
                    'fish_species': fish_species,
                    'collection_site': collection_site,
                    'video_id': video_id,
                    'frame_number': frame_num,
                    'npz_path': npz_path,
                    'model_confidence': candidate['confidence'],
                    'predicted_species': candidate['predicted_species'],
                    'is_correct': candidate['is_correct'],
                    'prediction_type': 'CORRECT'
                }
                all_extracted_frames.append((image, info))

        print(f"\n=== Extracting Incorrect Prediction Frames ===")
        for candidate in selected_incorrect:
            npz_path = candidate['npz_path']
            print(f"Processing {os.path.basename(npz_path)} (confidence: {candidate['confidence']:.3f})")

            image, fish_species, collection_site, video_id, frame_num = extract_frame_from_npz(npz_path)

            if image is not None:
                info = {
                    'fish_species': fish_species,
                    'collection_site': collection_site,
                    'video_id': video_id,
                    'frame_number': frame_num,
                    'npz_path': npz_path,
                    'model_confidence': candidate['confidence'],
                    'predicted_species': candidate['predicted_species'],
                    'is_correct': candidate['is_correct'],
                    'prediction_type': 'INCORRECT'
                }
                all_extracted_frames.append((image, info))

        print(f"\nSuccessfully extracted {len(all_extracted_frames)} frames total")
        print(f"- Correct: {sum(1 for _, info in all_extracted_frames if info['is_correct'])}")
        print(f"- Incorrect: {sum(1 for _, info in all_extracted_frames if not info['is_correct'])}")

        return all_extracted_frames

    except Exception as e:
        print(f"Error in frame extraction: {e}")
        return []

def run_task_level_occlusion_analysis(model, preprocess, device, data_dir, confidence_threshold=0.6,
                                     max_correct=5, max_incorrect=5, window_size=12, stride=4):
    """
    Main function to run task-level occlusion analysis on both correct and incorrect predictions
    """

    # 1. Load data and train SVM
    print("=== Training SVM Model ===")
    trained_svm, label_encoder, data_splits, X, y = load_data_and_train_svm(data_dir, seed=42)

    # 2. Extract both correct and incorrect frames
    print(f"\n=== Extracting Frames (threshold: {confidence_threshold}) ===")
    test_frames = extract_correct_and_incorrect_frames(
        data_dir, trained_svm, label_encoder, X, y,
        confidence_threshold=confidence_threshold,
        max_correct=max_correct, max_incorrect=max_incorrect, seed=42
    )

    if not test_frames:
        print("No frames found! Cannot proceed with analysis.")
        return

    # 3. Initialize task-level analyzer
    analyzer = TaskLevelOcclusionAnalyzer(
        model=model,
        preprocess=preprocess,
        device=device,
        trained_svm=trained_svm,
        label_encoder=label_encoder
    )

    # 4. Run task-level analysis on each frame
    print(f"\n=== Running Task-Level Analysis on {len(test_frames)} Frames ===")
    results = []

    for i, (image, info) in enumerate(test_frames, 1):
        print(f"\n{'='*60}")
        print(f"FRAME {i}/{len(test_frames)} - {info['prediction_type']}")
        print(f"{'='*60}")
        print(f"Actual Species: {info['fish_species']}")
        print(f"Predicted Species: {info['predicted_species']}")
        print(f"Model Confidence: {info['model_confidence']:.3f}")
        print(f"Correct Prediction: {info['is_correct']}")
        print(f"Video: {info['collection_site']}_{info['video_id']}, Frame: {info['frame_number']}")

        # Save with descriptive name including correctness
        confidence_str = f"{info['model_confidence']:.3f}".replace('.', 'p')
        correctness = "CORRECT" if info['is_correct'] else "INCORRECT"
        save_name = f"task_level_{correctness}_conf_{confidence_str}_{info['collection_site']}_{info['video_id']}_{info['predicted_species']}.png"
        save_path = f"/content/{save_name}"

        # Run task-level occlusion analysis
        result = analyzer.task_level_occlusion_analysis(
            image,
            save_path=save_path,
            window_size=window_size,
            stride=stride
        )

        # Add metadata to result
        result['info'] = info
        results.append(result)

    # 5. Summary analysis
    print(f"\n{'='*60}")
    print("SUMMARY OF TASK-LEVEL ANALYSIS")
    print(f"{'='*60}")

    correct_results = [r for r in results if r['info']['is_correct']]
    incorrect_results = [r for r in results if not r['info']['is_correct']]

    print(f"Analyzed {len(correct_results)} correct predictions and {len(incorrect_results)} incorrect predictions")

    if correct_results:
        print(f"\nCorrect Predictions:")
        for r in correct_results:
            print(f"- {r['info']['predicted_species']}: confidence = {r['confidence']:.3f}")

    if incorrect_results:
        print(f"\nIncorrect Predictions:")
        for r in incorrect_results:
            print(f"- {r['info']['fish_species']} → {r['info']['predicted_species']}: confidence = {r['confidence']:.3f}")

    return results

# USAGE FUNCTIONS

def main_task_level_analysis(model, preprocess, device, npz_dir, confidence_threshold=0.6,
                            max_correct=5, max_incorrect=5, window_size=12, stride=4):
    """
    Main function for task-level occlusion analysis on both correct and incorrect predictions

    Args:
        model: SigLIP model
        preprocess: Preprocessing function
        device: Device (cuda/cpu)
        npz_dir: Directory containing NPZ files
        confidence_threshold: Minimum confidence for frame selection
        max_correct: Maximum number of correct predictions to analyze
        max_incorrect: Maximum number of incorrect predictions to analyze
        window_size: Size of occlusion window
        stride: Step size for occlusion window

    Usage:
        results = main_task_level_analysis(model, preprocess, device, "/path/to/npz/files")
    """
    print("Starting Task-Level Occlusion Analysis...")
    print(f"Confidence threshold: {confidence_threshold}")
    print(f"Max correct predictions: {max_correct}")
    print(f"Max incorrect predictions: {max_incorrect}")
    print(f"Window size: {window_size}x{window_size}, stride: {stride}")
    print(f"Data directory: {npz_dir}")

    results = run_task_level_occlusion_analysis(
        model=model,
        preprocess=preprocess,
        device=device,
        data_dir=npz_dir,
        confidence_threshold=confidence_threshold,
        max_correct=max_correct,
        max_incorrect=max_incorrect,
        window_size=window_size,
        stride=stride
    )

    return results

# Example usage:
# Clear hooks first
clear_all_hooks(model)

# Run task-level analysis
npz_dir = "/content/drive/MyDrive/Summer 2023 - fish ML/Paper - Fish Species Classification/Feature Extraction/ViT-SO400M-14-SigLIP"
results = main_task_level_analysis(model, preprocess, device, npz_dir)

print("Task-level only occlusion analysis code loaded successfully!")
print("\nTo use this code:")
print("1. Clear any existing hooks: clear_all_hooks(model)")
print("2. Run analysis: results = main_task_level_analysis(model, preprocess, device, npz_dir)")
print("\nThis will:")
print("- Train an SVM on your data")
print("- Extract high-confidence frames (both correct and incorrect)")
print("- Run ONLY task-level (SVM) occlusion analysis")
print("- Create visualizations showing original image + task-level importance")
print("- Use smaller colorbar for cleaner appearance")
```
